# YOLOv8n vs Best.pt ÏÑ±Îä• ÎπÑÍµê

```python
# ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
# Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞
model_base = YOLO("yolov8n.pt")   # COCO ÏÇ¨Ï†ÑÌïôÏäµ Î™®Îç∏
model_best = YOLO("runs/detect/train/weights/best.pt")  # ÎÇ¥Í∞Ä ÌïôÏäµÌïú best Î™®Îç∏
```

```python
# ÎπÑÍµêÌï† Ïù¥ÎØ∏ÏßÄ Î¶¨Ïä§Ìä∏ Ï†ïÏùò
images = [
    "mnist/000001.jpg",   # Ïà´Ïûê
    "cat.png",            # ÎèôÎ¨º
    "people.jpg",         # ÏÇ¨Îûå
    "car.png",            # ÏûêÎèôÏ∞®
    "sample.jpg"          # Í∞ÄÏúÑÎ∞îÏúÑÎ≥¥
]
```

```python
# ÎπÑÍµê Ìï®Ïàò Ï†ïÏùò
def compare_models(img_path):
    # ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ ÏùΩÍ∏∞
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Îëê Î™®Îç∏ ÏòàÏ∏°
    results_base = model_base.predict(img_path, verbose=False)
    results_best = model_best.predict(img_path, verbose=False)

    # Í≤∞Í≥º Ïù¥ÎØ∏ÏßÄ Í∞ÄÏ†∏Ïò§Í∏∞
    img_base = results_base[0].plot()
    img_best = results_best[0].plot()

    # BGR ‚Üí RGB Î≥ÄÌôò
    img_base = cv2.cvtColor(img_base, cv2.COLOR_BGR2RGB)
    img_best = cv2.cvtColor(img_best, cv2.COLOR_BGR2RGB)

    # ÎπÑÍµê Ï∂úÎ†•
    fig, axes = plt.subplots(1, 3, figsize=(15, 6))
    axes[0].imshow(img_rgb)
    axes[0].set_title("Original")
    axes[1].imshow(img_base)
    axes[1].set_title("YOLOv8n (Pretrained)")
    axes[2].imshow(img_best)
    axes[2].set_title("Best.pt (Custom)")
    for ax in axes:
        ax.axis("off")
    plt.show()
```

```python
# Ïù¥ÎØ∏ÏßÄÎ≥Ñ ÎπÑÍµê Ïã§Ìñâ
for img in images:
    compare_models(img)
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# 1. ÌïôÏäµÌïú best Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞
model = YOLO("runs/detect/train/weights/best.pt")

# 2. ÌÖåÏä§Ìä∏ Ïù¥ÎØ∏ÏßÄ Í≤ΩÎ°ú
img_path = "sample.png"   # Í∞ÄÏúÑÎ∞îÏúÑÎ≥¥

# 3. ÏòàÏ∏° Ïã§Ìñâ
results = model.predict(img_path, save=True, project="runs/test", name="rps_test")

# 4. Í≤∞Í≥º Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
res_img = results[0].plot()                # Î∞îÏö¥Îî©Î∞ïÏä§ Í∑∏Î†§ÏßÑ Ïù¥ÎØ∏ÏßÄ
res_img = cv2.cvtColor(res_img, cv2.COLOR_BGR2RGB)

plt.imshow(res_img)
plt.axis("off")
plt.show()

# 5. ÎùºÎ≤®, Ïã†Î¢∞ÎèÑ Ï∂úÎ†•
for box in results[0].boxes:
    cls_id = int(box.cls[0])
    conf = float(box.conf[0])
    print(f"Class: {model.names[cls_id]}, Confidence: {conf:.4f}")
```
