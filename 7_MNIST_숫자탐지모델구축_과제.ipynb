```python
import matplotlib.pyplot as plt

# 1. COCO 사전학습 모델 불러오기 (기본 yolov8n)
model_base = YOLO("yolov8n.pt")

# 2. 손글씨 숫자 데이터셋으로 전이학습
# 결과는 runs/detect/yolov8n_numbers/weights/ 에 저장됨
model = YOLO("yolov8n.pt")
results = model.train(
    data="./numbers/data.yaml",  # 데이터셋 경로
    epochs=50,                   # 학습 반복 횟수
    imgsz=640,                   # 이미지 크기
    batch=16,                    # 배치 크기
    name="yolov8n_numbers"       # 결과 폴더명
)

# 3. 학습된 모델(best.pt) 불러오기
model_best = YOLO("runs/detect/yolov8n_numbers/weights/best.pt")

# 4. 성능 평가 (Validation dataset 기준)
metrics_base = model_base.val(data="./numbers/data.yaml")
metrics_best = model_best.val(data="./numbers/data.yaml")

# 5. 성능 비교 시각화
labels = [
    'metrics/precision(B)',
    'metrics/recall(B)',
    'metrics/mAP50(B)',
    'metrics/mAP50-95(B)'
]
plot_labels = ['precision', 'recall', 'mAP50', 'mAP50-95']  # 그래프 축에 표시할 이름

base_scores = [metrics_base.results_dict[m] for m in labels]
best_scores = [metrics_best.results_dict[m] for m in labels]

x = range(len(labels))
plt.bar(x, base_scores, width=0.4, label='Base (yolov8n pretrained)', align='center')
plt.bar([i+0.4 for i in x], best_scores, width=0.4, label='Fine-tuned (yolov8n_numbers)', align='center')
plt.xticks([i+0.2 for i in x], plot_labels)
plt.ylabel("Score")
plt.title("YOLOv8n vs YOLOv8n_numbers (Fine-tuned)")
plt.legend()
plt.show()
```

```python
# 비교할 지표 키
keys = [
    'metrics/precision(B)',
    'metrics/recall(B)',
    'metrics/mAP50(B)',
    'metrics/mAP50-95(B)',
    'fitness'
]

print("=== 성능 지표 비교 ===")
for k in keys:
    base_val = metrics_base.results_dict.get(k, None)
    best_val = metrics_best.results_dict.get(k, None)
    print(f"{k:<20} | Base: {base_val:.4f} | Fine-tuned: {best_val:.4f}")
```
