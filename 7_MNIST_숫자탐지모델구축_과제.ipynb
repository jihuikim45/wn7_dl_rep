```python
from ultralytics import YOLO

# ÏÜêÍ∏ÄÏî® Ïà´Ïûê Îç∞Ïù¥ÌÑ∞ÏÖãÏúºÎ°ú Ï†ÑÏù¥ÌïôÏäµ
model.train(
    data="./numbers/data.yaml",
    epochs=50,
    imgsz=640,
    name="yolov8n_numbers"   # <-- Í≤∞Í≥º Ìè¥Îçî Ïù¥Î¶Ñ
)
```

```python
import matplotlib.pyplot as plt

# 1. COCO ÏÇ¨Ï†ÑÌïôÏäµ Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞ (Í∏∞Î≥∏ yolov8n)
model_base = YOLO("yolov8n.pt")

# 2. ÏÜêÍ∏ÄÏî® Ïà´Ïûê Îç∞Ïù¥ÌÑ∞ÏÖãÏúºÎ°ú Ï†ÑÏù¥ÌïôÏäµ
# Í≤∞Í≥ºÎäî runs/detect/yolov8n_numbers/weights/ Ïóê Ï†ÄÏû•Îê®
model = YOLO("yolov8n.pt")
results = model.train(
    data="./numbers/data.yaml",  # Îç∞Ïù¥ÌÑ∞ÏÖã Í≤ΩÎ°ú
    epochs=50,                   # ÌïôÏäµ Î∞òÎ≥µ ÌöüÏàò
    imgsz=640,                   # Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞
    batch=16,                    # Î∞∞Ïπò ÌÅ¨Í∏∞
    name="yolov8n_numbers"       # Í≤∞Í≥º Ìè¥ÎçîÎ™Ö
)

# 3. ÌïôÏäµÎêú Î™®Îç∏(best.pt) Î∂àÎü¨Ïò§Í∏∞
model_best = YOLO("runs/detect/yolov8n_numbers/weights/best.pt")

# 4. ÏÑ±Îä• ÌèâÍ∞Ä (Validation dataset Í∏∞Ï§Ä)
metrics_base = model_base.val(data="./numbers/data.yaml")
metrics_best = model_best.val(data="./numbers/data.yaml")

# 5. ÏÑ±Îä• ÎπÑÍµê ÏãúÍ∞ÅÌôî
labels = [
    'metrics/precision(B)',
    'metrics/recall(B)',
    'metrics/mAP50(B)',
    'metrics/mAP50-95(B)'
]
plot_labels = ['precision', 'recall', 'mAP50', 'mAP50-95']  # Í∑∏ÎûòÌîÑ Ï∂ïÏóê ÌëúÏãúÌï† Ïù¥Î¶Ñ

base_scores = [metrics_base.results_dict[m] for m in labels]
best_scores = [metrics_best.results_dict[m] for m in labels]

x = range(len(labels))
plt.bar(x, base_scores, width=0.4, label='Base (yolov8n pretrained)', align='center')
plt.bar([i+0.4 for i in x], best_scores, width=0.4, label='Fine-tuned (yolov8n_numbers)', align='center')
plt.xticks([i+0.2 for i in x], plot_labels)
plt.ylabel("Score")
plt.title("YOLOv8n vs YOLOv8n_numbers (Fine-tuned)")
plt.legend()
plt.show()
```

```python
# ÎπÑÍµêÌï† ÏßÄÌëú ÌÇ§
keys = [
    'metrics/precision(B)',
    'metrics/recall(B)',
    'metrics/mAP50(B)',
    'metrics/mAP50-95(B)',
    'fitness'
]

print("=== ÏÑ±Îä• ÏßÄÌëú ÎπÑÍµê ===")
for k in keys:
    base_val = metrics_base.results_dict.get(k, None)
    best_val = metrics_best.results_dict.get(k, None)
    print(f"{k:<20} | Base: {base_val:.4f} | Fine-tuned: {best_val:.4f}")
```
